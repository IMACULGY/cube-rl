```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

## Loss function

```python
# get numpy array for API and dqn (NO HUMAN KNOWLEDGE)
losses_api = np.load('api_loss.npy')
losses_dqn = np.load('dqn_loss.npy')

# get numpy array for API and dqn (WITH HUMAN KNOWLEWDGE)
losses_api_h = np.load('api_loss_human.npy')
losses_dqn_h = np.load('dqn_loss_human.npy')
```

```python
# plot the losses side-by-side

plt.subplot(2, 2, 1)
plt.xlabel('iteration')
plt.ylabel('training loss')
plt.title('Loss (no human)')
l1, =plt.plot(losses_api, label = 'API')
l2, =plt.plot(losses_dqn, label = 'DQN')
plt.legend(handles=[l1,l2])


plt.subplot(2, 2, 2)
plt.xlabel('iteration')
plt.ylabel('training loss')
plt.title('Loss (human)')
l1, =plt.plot(losses_api_h, label = 'API')
l2, =plt.plot(losses_dqn_h, label = 'DQN')
plt.legend(handles=[l1,l2])
```
